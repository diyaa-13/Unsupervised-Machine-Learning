```python
#Digit Dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler , LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn import metrics
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
digits
```

```python
x = digits.data
y = digits.target
```

```python
x
```

```python
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
clf = RandomForestClassifier(n_estimators=100,random_state=42)
clf.fit(x_train_scaled, y_train)
```

```python
y_pred = clf.predict(x_test_scaled)
print("Accuracy on Raw Data:",metrics.accuracy_score(y_test,y_pred))
```

```python
pca = PCA(n_components=20)
x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.transform(x_test_scaled)
```

```python
clf_pca = RandomForestClassifier(n_estimators=100,random_state=42)
clf_pca.fit(x_train_pca,y_train)
```

```python
y_pred_pca = clf_pca.predict(x_test_pca)
print("Accuracy with PCS (20 components): ",metrics.accuracy_score(y_test,y_pred_pca))
```

```python
#Using KNN Algorithm
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled, y_train)
```

```python
y_pred_knn = knn.predict(x_test_scaled)
print("Accuracy with Raw Data:", metrics.accuracy_score(y_test, y_pred_knn))
```

```python
knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(x_train_pca, y_train)
```

```python
y_pred_knn_pca = knn_pca.predict(x_test_pca)
print("Accuracy with PCA :", metrics.accuracy_score(y_test, y_pred_knn_pca))
```

```python
print("Accuracy with Raw Data:", metrics.accuracy_score(y_test, y_pred_knn))
print("Accuracy with PCA :", metrics.accuracy_score(y_test, y_pred_knn_pca))
```
